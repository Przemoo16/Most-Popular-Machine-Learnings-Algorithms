# K-Means Clustering - easier

## Prepare data

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Generate data
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
#Plot generated data
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

## K Means Clustering

```python
#Create KMeans for 4 clusters
kmeans = KMeans (n_clusters=4)
```

```python
#Train classifier
kmeans.fit(data[0])
```

```python
#Co-ordinates of each cluster
kmeans.cluster_centers_
```

```python
#Cluster for each data point
kmeans.labels_
```

```python
#Create figure for subplot
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

#Plot data grouped by K Means
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

#Plot original data
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')

#The order of colors is not important. It is important what samples have been assigned to a particular cluster.
```
